```
import pandas as pd
import ast
import numpy as np

final_papers = pd.read_csv('papers.csv')
final_authors = pd.read_csv('authors.csv')
```

```
final_authors['H_index'] = final_authors['papers']
final_authors['H_index'] = final_authors.papers.apply(lambda x:  list(map(np.int64, ast.literal_eval(x))))
final_authors['H_index'] = final_authors.H_index.apply(lambda x: [final_papers.loc[final_papers['id'] == j]['n_citation'].tolist()[0] for j in x])

final_authors['H_index'] = final_authors.H_index.apply(lambda x: sum(j >= i + 1 for i, j in enumerate(sorted(list(x), reverse=True))))


final_authors.to_csv('authors.csv')
```
