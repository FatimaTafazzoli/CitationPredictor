```
import pandas as pd
import ast 

count = 0
table = []
papers = []
author = []
orgs = []
orgsName = []
ids = []
```

```
with open ('dblp_papers_v11.txt','r') as f:
  line = f.readline()  
  while line:
    data = ast.literal_eval(line)

    if 'volume' in data.keys():
        del data['volume']
    if 'issue' in data.keys():
        del data['issue']
    if 'fos' in data.keys():
        del data['fos']
    if 'indexed_abstract' in data.keys():
        del data['indexed_abstract']
    
    pids = []
    data['first_author'] = data['authors'][0]['id']
    for i in data['authors']:
        pids.append(i['id'])
        if i['id'] in ids:
            author[ids.index(i['id'])]['papers'].append(data['id'])
        else:
            author.append(i)
            ids.append(i['id'])
            author[-1]['papers']=[data['id']]
            
        if 'org' in i.keys():
           if i['org'] in orgsName:
               orgs[orgsName.index(i['org'])]['authors'].append(i['id'])
               orgs[orgsName.index(i['org'])]['papers'].append(data['id'])
               orgs[orgsName.index(i['org'])]['papers'] = list(set(orgs[orgsName.index(i['org'])]['papers']))
               orgs[orgsName.index(i['org'])]['authors'] = list(set(orgs[orgsName.index(i['org'])]['authors']))
           else:
               orgsName.append(i['org'])
               org = {'name':i['org'], 'authors': [i['id']], 'papers': [data['id']]}
               orgs.append(org)

    data['authors'] = pids
    papers.append(data)
    
    line = f.readline()
    count+=1
```

```
final_papers = pd.DataFrame(papers, columns =('id', 'title', 'authors', 'venue', 'year', 'n_citation', 'page_start', 'page_end', 'doc_type', 'publisher', 'doi', 'references','first_author'))
final_authors = pd.DataFrame(author, columns =('name', 'id', 'org', 'papers'))
final_orgs = pd.DataFrame(orgs, columns =('name', 'authors', 'papers'))


final_papers.to_csv('papers.csv')
final_authors.to_csv('authors.csv')
final_orgs.to_csv('orgs.csv')
```
